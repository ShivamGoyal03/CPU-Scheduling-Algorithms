```python
import threading
import time
import random

# Number of philosophers
N = 5

# Create a list of fork locks
forks = [threading.Lock() for _ in range(N)]

# Create a list to track which philosophers have eaten
philosophers_eaten = [False] * N

# Create a counter to keep track of how many philosophers have eaten
philosophers_eaten_count = 0

def philosopher(i):
    global philosophers_eaten_count

    while philosophers_eaten_count < N:
        if not philosophers_eaten[i]:
            print(f'Philosopher {i} is thinking.')
            time.sleep(random.uniform(0, 1))

            print(f'Philosopher {i} is hungry.')
            # Acquire the left fork
            forks[i].acquire()
            print(f'Philosopher {i} has acquired the left fork.')

            # Attempt to acquire the right fork
            right = (i + 1) % N
            if forks[right].acquire(blocking=False):
                print(f'Philosopher {i} has acquired the right fork and is eating.')
                time.sleep(random.uniform(1, 3))
                forks[right].release()
                print(f'Philosopher {i} has released the right fork.')

                # Mark the philosopher as having eaten
                philosophers_eaten[i] = True
                print(f'Philosopher {i} has eaten.')

                # Increment the count of philosophers who have eaten
                philosophers_eaten_count += 1

            # Release the left fork
            forks[i].release()
            print(f'Philosopher {i} has released the left fork.')

if __name__ == '__main__':
    philosophers = [threading.Thread(target=philosopher, args=(i,)) for i in range(N)]

    for philosopher in philosophers:
        philosopher.start()

    for philosopher in philosophers:
        philosopher.join()
```
